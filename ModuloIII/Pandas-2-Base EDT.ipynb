#DIY
- crie uma nova coluna categorica com a coluna run time com as opções curta_duracao(30<), media_duracao(entre 30 e 90), longa_duracao(>90)
- faça um group by com essas nova coluna 
- atribua cada grupo a uma variavel escolha so um type = SHOW ou MOVIE
- veja o tamanho do dataframe e defina quantos filmes tem cada categoria nova criada 
- pegue o top 10 de cada grupo (imdb score)

```python
df_1['duracao'] = df_1['runtime'].apply(lambda x: 'curta_duracao' if x < 30 else 'media_duracao' if 30 <= x <= 90 else 'longa_duracao')
```

```python
df_1
```

```python
df_1.duracao = pd.Categorical(df_1.duracao, ['curta_duracao', 'media_duracao', 'longa_duracao'], ordered=True)
```
