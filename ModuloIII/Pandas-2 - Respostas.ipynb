#DIY
- crie uma nova coluna categorica com a coluna run time com as opções curta_duracao(30<), media_duracao(entre 30 e 90), longa_duracao(>90)
- faça um group by com essas nova coluna e com a coluna type
- atribua cada grupo a uma variavel escolha so um type - SHOW ou MOVIE
- veja o tamanho do dataframe e defina quantos filmes tem cada categoria nova criada 
- pegue o top 10 de cada grupo (imdb score)

```python
def run_time(x):
    if x <= 30:
        opcao = 'curta_duracao'
    elif x >= 30 and x<=90:
        opcao = 'media_duracao'
    elif x>90:
        opcao = 'longa_duracao'
    return opcao
```

```python
df_1['cat_runtime'] = df_1['runtime'].apply(lambda x: run_time(x))
```

```python
df_1.head()
```

```python
df_1_curta = df_1.groupby(['cat_runtime'])
```

```python
df_1_curta = dict(list(df_1.groupby(['cat_runtime'])))['curta_duracao']
df_1_media = dict(list(df_1.groupby(['cat_runtime'])))['media_duracao']
df_1_longa = dict(list(df_1.groupby(['cat_runtime'])))['longa_duracao']
```

```python
print(df_1_curta.shape)
print(df_1_media.shape)
print(df_1_longa.shape)
```

```python
df_1_curta.nlargest(10, 'imdb_score')
df_1_media.nlargest(10, 'imdb_score')
df_1_longa.nlargest(10, 'imdb_score')
```
