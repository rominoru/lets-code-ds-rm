#### Series 

```python
#Pandas Series
#Um grupo de series forma um dataframe
ds1 = pd.Series(np.random.randint(10, 25, size=25))
ds2 = pd.Series(np.random.randint(10, 25, size=25))
```

```python
print(type(ds1))
print(ds1.shape)
print(ds1)
```

```python
print(type(ds2))
print(ds2.shape)
print(ds2)
```

```python
#Matematica
ds = ds1 + ds2
print('Soma:')
print(ds)
print('\n')
print('Subtração:')
ds = ds1 - ds2
print(ds)
print('\n')
print('Multiplicação:')
ds = ds1 * ds2
print(ds)
print('\n')
print("Divisão:")
ds = ds1 / ds2
print(ds)
print('\n')
print('Media:')
print(ds1.mean())
print('\n')
print('Mediana:')
print(ds1.median())
print('\n')
print('Std Dev:')
print(ds1.std())
print('\n')
print('\n')
```

```python
#Comparando series
print("Igual:")
print(ds1 == ds2)
print("Maior que:")
print(ds1 > ds2)
print("Menor que:")
print(ds1 < ds2)
```

##### DYI Series

- Crie duas series aleatorias com 30 numeros, no intervalo 20 a 50 cada  ds1 ds2
- Crie uma nova serie com o nome g_than com a comparação entre a s1 e a s2 
- Crie uma nova serie com o nome ds_sum com a soma das duas series 
- Apresente o desvio padrão da serie ds_sum
- Crie uma nova serie com 30 strings de 5 caracteres aleatorios (Dica. random.choice + string.ascii_lowercase)
- Crie uma lista com todas as series criadas 
- Crie um dataframe com as suas series (Dica. pd.concat())
- Coloque o nome de cada serie no novo dataframe (Dica .rename)

```python
import random 
import string

ds1 = pd.Series(np.random.randint(20, 51, size=30))
ds2 = pd.Series(np.random.randint(20, 51, size=30))
# ds1, ds2
```

```python
g_than = pd.Series(ds1>ds2)
```

```python
ds_sum = pd.Series(ds1+ds2)
# ds_sum
```

```python
ds_sum.std()
```

```python
def get_random_str():
    return ''.join([random.choice(string.ascii_lowercase) for i in range(5)])
    
rand_strs = [get_random_str() for i in range(30)]
```

```python
rand_strs
```

```python
ds3 = pd.Series(rand_strs)
# ds3
```

```python
print(type(ds1))
print(type(ds2))
print(type(g_than))
print(type(ds_sum))
print(type(ds3))
```

```python
# df = pd.concat([ds1.reset_index(drop=True),ds2.reset_index(drop=True),g_than.reset_index(drop=True),ds_sum.reset_index(drop=True),ds3.reset_index(drop=True)], axis=1)
df = pd.concat([ds1,ds2,g_than,ds_sum,ds3], axis=1)
df.rename(columns={
    0:'ds1',
    1:'ds2',
    2:'g_than',
    3:'ds_sum',
    4:'ds3'
},inplace=True)
df
```

```python
#Pandas as Type
#https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.astype.html
df_csv['Valor de Venda'] = df_csv['Valor de Venda'].apply(lambda x:x.replace(',','.'))

df_csv['Valor de Venda'] = df_csv['Valor de Venda'].astype('float')

```

>Uma função lambda do Python se comporta como uma função normal em relação aos argumentos. Portanto, um parâmetro lambda pode ser inicializado com um valor padrão: o parâmetro **n** toma o **n** externo como valor padrão.

```python
df_csv['Numero Rua'] = df_csv['Numero Rua'].apply(lambda x:0 if x == 'S/N' or 'SN' else x)
#As Type mais de uma coluna
df_csv = df_csv.astype({"Valor de Venda":float,"Numero Rua":int})
```
