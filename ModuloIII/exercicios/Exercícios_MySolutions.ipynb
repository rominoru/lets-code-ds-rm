```python
# Divida sua serie em duas partes e gere duas series novas 
pdSeries = pd.Series(np.random.randint(5,16,30))
pdSeries_1 = pdSeries[0:15].copy().reset_index(drop=True)
pdSeries_2 = pdSeries[15:30].copy().reset_index(drop=True)
```

```python
pdSeries
```

```python
pdSeries_1
```

```python
pdSeries_2
```

```python
# Adicione 5 valores a cada serie e mantenha os dois elementos como series
add_1 = pd.Series(np.random.randint(5,16,5))
add_2 = pd.Series(np.random.randint(5,16,5))
concatS1 = pd.concat([pdSeries_1, add_1]).reset_index(drop=True)
concatS2 = pd.concat([pdSeries_2, add_2]).reset_index(drop=True)
```

```python
concatS1
```

```python
concatS2
```

```python
# Conte as aparições de cada elemento da Serie
concatS1.value_counts()
```

```python
concatS2.value_counts()
```

```python
# Faça comparações booleanas entre as das series
filter_1 = concatS1 > concatS2
filter_2 = concatS1 < concatS2
filter_3 = concatS1 == concatS2
filter_4 = concatS1 != concatS2
```
