```python
import numpy as np
import random
```

```python
from random import randrange
from datetime import timedelta
from datetime import datetime

def random_date(start, end):
    delta = end - start
    int_delta = (delta.days * 24 * 60 * 60) + delta.seconds
    random_second = randrange(int_delta)
    return start + timedelta(seconds=random_second)
```

```python
max_users = 15
num_titles = 330

d1 = datetime.strptime('2020/1/1 12:01 AM', '%Y/%m/%d %I:%M %p')
d2 = datetime.strptime('2022/7/31 11:59 PM', '%Y/%m/%d %I:%M %p')
```

```python
# avaliações
ratings_list = [[(u, random.randint(1,num_titles), random.randint(1,5), str(random_date(d1, d2))) for i in range(random.randint(50,120))] for u in range(1,max_users+1)]
```

```python
str(ratings_list).replace('[','').replace(']','').replace('  ','')
```

```python
# historico acoes
hist_list = []
for user in range(1,max_users+1):
    for i in range(random.randint(50,120)):
        title_id = random.randint(1,num_titles)
        random_time = random_date(d1, d2)
        start_time = str(random_time)
        end_time = str(random_time + timedelta(minutes=random.randint(45,130)))
        hist_list.append((user, title_id, start_time, 1))
        hist_list.append((user, title_id, end_time, 0))
```
